```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

from assignment import FIST_image,FIST_features
```

### Loading images

```python
X = np.array(Image.open('Images/landscape4.jpeg'),dtype=float)[:,:,:3]
Y_ = Image.open('Images/landscape1.jpeg')

scale = 1.33
Y = Y_.resize((int(scale*Y_.width),int(scale*Y_.height)),0)
Y = np.array(Y,dtype=float)
```

### FIST algorithm

```python
n_iter = 300
X_match = FIST_image(X,Y,n_iter).clip(0,255)
```

Alternatively, you can use FIST_features(X,Y,n_iter,dim=3) which is a generalization for any number of channels

### Plot

```python
fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(15,15))

axes[0].imshow(X.astype(np.int))
axes[0].set_title("Original image")
axes[0].axis("off")

axes[1].imshow(X_match.astype(np.int))
axes[1].set_title("transformed image")
axes[1].axis("off")

axes[2].imshow(Y.astype(np.int))
axes[2].set_title("color model")
axes[2].axis("off")

plt.tight_layout()
```
